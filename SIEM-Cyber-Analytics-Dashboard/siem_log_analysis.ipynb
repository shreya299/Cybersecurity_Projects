# SIEM Log Analysis
Python-based security log parsing & analytics

```
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
df = pd.read_csv('logs.csv')

# Convert timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date
df['hour'] = df['timestamp'].dt.hour

# Failed Logons
failed = df[df['status'] == 'FAILED']
failed_by_user = failed.groupby('username').size()

# Export CSVs
failed.to_csv('failed_logons_export.csv', index=False)
df.to_csv('cleaned_logs_export.csv', index=False)

# Visualization
failed_by_user.plot(kind='bar', figsize=(6,4))
plt.title('Failed Logons by User')
plt.xlabel('User')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('failed_logons_chart.png')
plt.show()
```
